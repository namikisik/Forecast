```python
import pandas as pd
import numpy as np
import pandas_datareader.data as web
import datetime
import cufflinks as cf
import plotly as py
cf.go_offline()
```

```python
enddate = str(datetime.date.today()- datetime.timedelta(days=1))
symbols = {'Gold':'GC=F', 'USD_TL':'TRY=X', 'BIST30':'XU030.IS'}
#symbols = {'BIST30':'XU030.IS',
#            'THY':'THYAO.IS',
#            'PETKIM':'PETKM.IS',
#            'ASELSAN':'ASELS.IS',
#            'TUPRAS':'TUPRS.IS',
#            'SABANCI HOL':'SAHOL.IS',
#            'KOC HOL':'KCHOL.IS'}
```

```python
def period(days):
    days = str(datetime.date.today()- datetime.timedelta(days=days))
    return str(days)
```

```python
def ret(df):
    # calculate daily return
    df_daily= df.pct_change()
    # calculate cumluative return
    df_cumul = np.exp(np.log1p(df_daily).cumsum())
    return df_daily, df_cumul
```

```python
def test_run(days):
    for i in range(len(symbols)):
            temp = 'df_{}'.format(list(symbols)[i])
            vars()[temp] = web.DataReader(symbols[list(symbols)[i]], 'yahoo', start=period(days), end=enddate,)
            vars()[temp]['Daily'], vars()[temp]['Cumu'] = ret(vars()[temp]['Adj Close'])
            qf=cf.QuantFig(vars()[temp],title=list(symbols)[i] + ' Quant Figure',legend='top',name='GS')
            qf.add_bollinger_bands()
            qf.add_volume()

            qf.iplot()
```

```python
if __name__ == "__main__":
    test_run(200)
```

```python
for i in range(len(symbols)):
    temp = 'df_{}'.format(list(symbols)[i])
    vars()[temp] = web.DataReader(symbols[list(symbols)[i]], 'yahoo', start=period(200), end=enddate,)
    vars()[temp]['Daily'], vars()[temp]['Cumu'] = ret(vars()[temp]['Adj Close'])
    qf=cf.QuantFig(vars()[temp],title=list(symbols)[i])
    qf.add_sma([20,50],width=2,color=['green','lightgreen'],legendgroup=True)
    qf.add_rsi(periods=20,color='java')
    qf.add_bollinger_bands(periods=20,boll_std=2,colors=['magenta','grey'],fill=True)
    #qf.add_volume()
    qf.add_macd()
    fig = qf.iplot(theme='white', asFigure=True)
    py.offline.plot(fig, filename="%s_%s.%s" %(list(symbols)[i],'bollingerBand','html'), auto_open=False)
```
